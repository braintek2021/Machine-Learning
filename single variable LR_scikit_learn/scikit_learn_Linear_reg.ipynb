# Example problem with scikit-learn

# Step 1: Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
cd "F:\Knowledge\ML\courses_ML\Reference\ml-coursera-python-assignments-master\Exercise1"
```

# Step 2: Load the Data

```python
# Read comma separated data
data = np.loadtxt(os.path.join('Data', 'ex1data1.txt'), delimiter=',')
X, y = data[:, 0], data[:, 1]
m = y.size  # number of training examples
#print(X.shape)
```

# Step 3: Plot Data

```python
print('Plotting Data ...\n')
def plotData(x, y):
    plt.plot(x, y, 'ro', ms=10, mec='k')
    plt.ylabel('Profit in $10,000')
    plt.xlabel('Population of City in 10,000s')
    fig = plt.figure()
```

```python
plotData(X, y)
```

# Step 4: Estimating Coefficients

```python
#sklearn.linear_model.LinearRegression help
#https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression
```

```python
print("The first five samples {}".format(X[:5]))
print("The first five targets {}".format(y[:5]))
X= X.reshape(-1, 1)
y= y.reshape(-1, 1)
print("The number of samples in train set is {}".format(X.shape))
print("The number of samples in test set is {}".format(y.shape))
lr = LinearRegression()
lr.fit(X, y)
```

# Step 5:  Prediction & Model Evaluation

```python
y_pred = lr.predict(X)
plt.scatter(X, y, color ='b')
plt.plot(X, y_pred, color ='k')
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
test_x= np.array([[35000]])
#test_x= test_x.reshape(-1, 1)
Pred_y = lr.predict(test_x)
print(Pred_y)
```

```python
#Y=mx+c
pred_y1 = 35000*1.19303364-3.89578088
print(pred_y1)
```

# Is this prediction good??

```python
print(lr.score(X, y))
```
