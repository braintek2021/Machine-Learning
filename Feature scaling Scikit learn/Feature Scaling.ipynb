# What is Feature Scaling - Normalization and Standardization
Scaling or Feature Scaling is the process of changinng the scale of certain features to a common one. This is typically achieved through normalization and standardization (scaling techniques).

Normalization is the process of scaling data into a range of [0, 1]. It's more useful and common for regression tasks.


Standardization is the process of scaling data so that they have a mean value of 0 and a standard deviation of 1. It's more useful and common for classification tasks

# Standardization with Python

```python
def Standardization(X):
    mean=np.mean(X,axis=0)
    std=np.std(X,axis=0)
    
    X_std = (X - mean)/std
    
    return X_std
```

```python
Y = Standardization(y)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6, 5))
ax1.set_title('Before standardization')
sns.kdeplot(y, ax=ax1)
ax2.set_title('After manual standardization')
sns.kdeplot(Y, ax=ax2)
plt.show()
```

# Normalization with python

```python
def normalize(X):
    Xmin=np.min(X)
    Xmax=np.max(X)
    
    X_norm = (X - Xmin)/(Xmax-Xmin)
    
    return X_norm
```

```python
Y_norm = normalize(y)
Y_norm=Y_norm.flatten()
```

# Normalization with sklearn

```python
from sklearn.preprocessing import MinMaxScaler
y= y.reshape(-1, 1)
scaler = MinMaxScaler()
Y_minmax = scaler.fit_transform(y)
Y_minmax=Y_minmax.flatten()
print(Y_norm.shape,Y_minmax.shape)
```
